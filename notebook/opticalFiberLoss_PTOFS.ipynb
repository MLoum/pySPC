# Calcul de l'ordre de grandeur des perts dans la fibre longue

Un fit par une loi en $1/\lambda^4$ sur les donées constructeur de la fibre GI50 (de chez sedi fibre) donne que l'atténuation $\alpha$ en db/km s'exprime en fonction de la longueur d'onde $\lambda$ en nl par :
$
\alpha = C +  K \times \dfrac{1}{\lambda^4}
$
avec $C = 0.09$ et $A=1.35E12$.  

D'autre part les pertes en db/km ont pour expression :
$
\alpha = -\dfrac{10}{L[km]} \log \left( \dfrac{P(L)}{P(0)} \right)
$
et donc :
$
 \dfrac{P(L)}{P(0)} = 10^{-\alpha L/10}
$

```python
import numpy as np
import matplotlib.pyplot as plt

wl = np.arange(0.4,0.8,0.001)
A = 1.3479
C = 0.089
L =0.1

alpha = C + A/(wl**4)


plt.plot(wl, alpha)
plt.xlabel("wavelength µm")
plt.ylabel("Attenuation")
plt.show()

transmission = np.power(10, -alpha * L/10)
plt.plot(wl, transmission)
plt.show()
```
